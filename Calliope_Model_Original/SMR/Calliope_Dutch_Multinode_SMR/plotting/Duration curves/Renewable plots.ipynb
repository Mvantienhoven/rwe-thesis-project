```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Renewables.ninja capacity factors/wind_nuts2_2050.csv"

data = pd.read_csv(file_path, parse_dates=["time"])

data = data.sort_values("time")

duration_curves = {}
nuts2_regions = [col for col in data.columns[1:] if col != "NL_TOTAL"]  # Exclude "NL_TOTAL"

for region in nuts2_regions:
    sorted_capacity = data[region].sort_values(ascending=False).reset_index(drop=True)
    duration_curves[region] = sorted_capacity

plt.figure(figsize=(12, 8))

for region, sorted_capacity in duration_curves.items():
    plt.plot(
        sorted_capacity.values,
        label=region,
        linewidth=2,
        alpha=0.8
    )

plt.xlabel("Hours per year", fontsize=20)
plt.ylabel("Capacity factor", fontsize=20)
plt.title("Capacity duration curves for wind", fontsize=22)
plt.legend(title="NUTS2 region", fontsize=16, loc="upper right", bbox_to_anchor=(1.2, 1))
plt.grid(True, linestyle="--", alpha=0.7)

# Save and show the plot
plt.tight_layout()
plt.savefig("Capacity_duration_curves_wind.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
