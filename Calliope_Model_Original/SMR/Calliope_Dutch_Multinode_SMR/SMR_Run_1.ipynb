```python
import calliope
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
import geopandas as gpd
import xarray as xr
import matplotlib.colors as mcolors
import matplotlib.ticker as mtick  # For percentage formatting
```

```python
calliope.set_log_verbosity("INFO", include_solver_output=True) 

# Loading the model
model = calliope.Model("SMR_model.yaml",scenario="capacities_2040,links_TenneT,demand_2040_nl")

model.run()
model_data = model._model_data

# Remove encoding for all variables
for var in model_data.variables:
    model_data[var].encoding = {}

# Save using xarray, specifying engine and format
model_data.to_netcdf("model_results_SMR_0.nc", engine="netcdf4", format="NETCDF4")
```

```python
import os
from datetime import datetime

def get_versioned_filename(base_name, ext, folder='.'):
    """Generate a filename with a timestamp to avoid overwriting."""
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    filename = f"{base_name}_{timestamp}.{ext}"
    full_path = os.path.join(folder, filename)
    return full_path

# Usage example:
versioned_file = get_versioned_filename('Model_1_run', 'html')
model.plot.summary(to_file=versioned_file)
print(f"Saved summary to {versioned_file}")
```
