```python
from pathlib import Path
import pandas as pd

# Define the base directory and build the full path
base_dir = Path.cwd()
file_dir = (
    base_dir
    / "Research_Runs"
    / "timeseries_data"
    / "demand"
    / "NBNL_EV_2040"
    / "PtH_Demand_EV.csv"
)

# Safely read the CSV file
if not file_dir.exists():
    raise FileNotFoundError(f"File not found: {file_dir}")

df = pd.read_csv(file_dir)

# Function to ensure each value is negative unless empty or already negative
def ensure_negative(val):
    # Skip NaN or blank entries
    if pd.isna(val) or str(val).strip() == "":
        return val
    val_str = str(val).strip()
    # Only prepend '-' if it isn't there yet
    if not val_str.startswith("-"):
        return f"-{val_str}"
    return val_str

# Apply to all columns except the first
df.iloc[:, 1:] = df.iloc[:, 1:].applymap(ensure_negative)

# Save the modified file back
df.to_csv(file_dir, index=False)
print("✅ Updated: all non-first-column values are negative where missing.")
```

```python
from pathlib import Path
import pandas as pd
import numpy as np

# 1. Define the base directory and build the full path
base_dir = Path.cwd()
file_dir = (
    base_dir
    / "Research_Runs"
    / "timeseries_data"
    / "demand"
    / "NBNL_EV_2040"
    / "EV_Power.csv"
)

# 2. Load the CSV file
df = pd.read_csv(file_dir)

# 3. Get all columns that can be interpreted as numeric
numeric_cols = df.columns[df.dtypes.apply(lambda dt: np.issubdtype(dt, np.number))]

print("="*70)
print("ORDER OF MAGNITUDE ANALYSIS")
print("="*70)

# 4. Analyze each numeric column
for col in numeric_cols:
    print(f"\n{'='*70}")
    print(f"COLUMN: {col}")
    print(f"{'='*70}")
    
    # Convert to numeric, coerce errors to NaN
    series = pd.to_numeric(df[col], errors='coerce')
    values = series.dropna().values  # drop NaN
    
    # Count zeros and non-zero
    zero_mask = values == 0
    zero_count = np.sum(zero_mask)
    non_zero_values = values[~zero_mask]
    non_zero_indices = np.where(~zero_mask)[0]
    
    total = len(values)
    print(f"Total values: {total}")
    print(f"Zero values: {zero_count} ({100 * zero_count / total:.2f}%)")
    print(f"Non-zero values: {len(non_zero_values)} ({100 * len(non_zero_values) / total:.2f}%)")
    
    if len(non_zero_values) == 0:
        print("No non-zero values to analyze.")
        continue
    
    # Calculate order of magnitude for each non-zero value
    abs_vals = np.abs(non_zero_values)
    orders = np.floor(np.log10(abs_vals)).astype(int)
    
    # Count frequency by order
    unique_orders, counts = np.unique(orders, return_counts=True)
    
    print(f"\nOrder of Magnitude Distribution:")
    print(f"{'Order':<6}  {'Range':<22}  {'Count':<8}  {'% Non-zero':<10}  {'Example Rows'}")
    print("-"*70)
    
    for order, count in zip(unique_orders, counts):
        pct = 100 * count / len(non_zero_values)
        lower = 10.0 ** order
        upper = 10.0 ** (order + 1)
        # Map orders back to original row indices
        example_rows = non_zero_indices[orders == order][:3]
        examples = ", ".join(map(str, example_rows))
        
        print(f"10^{order:<4}  [{lower:.2e} - {upper:.2e})  {count:<8}  {pct:>7.2f}%     Rows: {examples}")
    
    # Summary statistics
    print(f"\nSummary Statistics:")
    print(f"  Min abs value: {abs_vals.min():.6e} (10^{int(np.floor(np.log10(abs_vals.min())))})")
    print(f"  Max abs value: {abs_vals.max():.6e} (10^{int(np.floor(np.log10(abs_vals.max())))})")
    print(f"  Median:        {np.median(abs_vals):.6e} (10^{int(np.floor(np.log10(np.median(abs_vals))))})")
    print(f"  Mean:          {np.mean(abs_vals):.6e} (10^{int(np.floor(np.log10(np.mean(abs_vals))))})")

print(f"\n{'='*70}")
print("ANALYSIS COMPLETE")
print("="*70)
```

```python
from pathlib import Path
import pandas as pd
import numpy as np

# 1. Define base directory and build full path
base_dir = Path.cwd()
file_path = (
    base_dir
    / "Research_Runs"
    / "timeseries_data"
    / "renewables"
    / "wind_onshore_nodes_2040.csv"
)

# 2. Load CSV into DataFrame
df = pd.read_csv(file_path)

# 3. Identify columns to round (exclude time/index column)
#    Replace 'time' with your actual time column name if different
time_col = 'time'
cols_to_round = [c for c in df.columns if c != time_col]

# 4. Round up all values in these columns to 10^-3 (3 decimal places)
#    Using ceiling: np.ceil(x * 1e3) / 1e3
df[cols_to_round] = df[cols_to_round].applymap(
    lambda x: np.ceil(x * 1e3) / 1e3 if pd.notnull(x) else x
)

# 5. (Optional) Save the rounded DataFrame to a new CSV
output_path = file_path.with_name("wind_onshore_nodes_2040.csv")
df.to_csv(output_path, index=False)

print(f"Rounded values written to {output_path}")
```

```python
import yaml
from pathlib import Path
import numpy as np

# Define path to YAML file (adjust as needed)
base_dir = Path.cwd()
yaml_path = base_dir / "Research_Runs" / "model_config" / "scenarios" / "capacities_HI.yaml"

# Load YAML file
with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Extract all numeric values under 'overrides' -> scenario -> locations -> techs -> constraints
values = []
locations_info = []  # To store scenario, loc, tech, key for each value

for scenario, scenario_data in data.get('overrides', {}).items():
    locations = scenario_data.get('locations', {})
    for loc, loc_data in locations.items():
        techs = loc_data.get('techs', {})
        for tech, tech_data in techs.items():
            constraints = tech_data.get('constraints', {})
            for key, v in constraints.items():
                if isinstance(v, (float, int)) and v != 0:
                    values.append(v)
                    locations_info.append((scenario, loc, tech, key))

values_array = np.array(values)

if len(values_array) == 0:
    print("No non-zero numeric constraint values found.")
else:
    abs_vals = np.abs(values_array)
    orders = np.floor(np.log10(abs_vals)).astype(int)

    unique_orders, counts = np.unique(orders, return_counts=True)

    print("Order of Magnitude Distribution:")
    print(f"{'Order':<6}  {'Range':<22}  {'Count':<8}  {'Percentage':<10}")
    print("-" * 50)
    total = len(values_array)
    for order, count in zip(unique_orders, counts):
        pct = 100 * count / total
        lower = 10.0 ** order
        upper = 10.0 ** (order + 1)
        print(f"10^{order:<4}  [{lower:.2e} - {upper:.2e})  {count:<8}  {pct:>7.2f}%")

    print("\nSummary Statistics:")
    print(f"  Min abs value: {abs_vals.min():.6e} (10^{int(np.floor(np.log10(abs_vals.min())))})")
    print(f"  Max abs value: {abs_vals.max():.6e} (10^{int(np.floor(np.log10(abs_vals.max())))})")
    print(f"  Median:        {np.median(abs_vals):.6e} (10^{int(np.floor(np.log10(np.median(abs_vals))))})")
    print(f"  Mean:          {np.mean(abs_vals):.6e} (10^{int(np.floor(np.log10(np.mean(abs_vals))))})")

    # Find and print values with orders of magnitude -6 and -5 (e-6 and e-5)
    target_orders = [-6, -5]
    for target in target_orders:
        print(f"\nValues with order of magnitude 10^{target}:")
        found = False
        for idx, order in enumerate(orders):
            if order == target:
                scenario, loc, tech, key = locations_info[idx]
                val = values_array[idx]
                print(f"  Scenario: {scenario}, Location: {loc}, Tech: {tech}, Constraint: {key}, Value: {val:.6e}")
                found = True
        if not found:
            print(f"  No values found with order 10^{target}.")
```

```python
import yaml
from pathlib import Path
import numpy as np

# Define path to YAML file (adjust as needed)
base_dir = Path.cwd()
yaml_path = base_dir / "Research_Runs" / "model_config" / "scenarios" / "capacities_HI.yaml"

# Load YAML file
with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Initialize a list to store (scenario, location, tech, key) for each value
locations_info = []

# Modify to set small values to zero
for scenario, scenario_data in data.get('overrides', {}).items():
    locations = scenario_data.get('locations', {})
    for loc, loc_data in locations.items():
        techs = loc_data.get('techs', {})
        for tech, tech_data in techs.items():
            constraints = tech_data.get('constraints', {})
            for key, v in constraints.items():
                if isinstance(v, (float, int)):
                    if abs(v) < 1e-4:
                        constraints[key] = 0  # Set to zero
                    # Save info for reference if needed
                    if abs(v) >= 1e-4:
                        locations_info.append((scenario, loc, tech, key, v))

# Optionally, save the modified data back to the YAML file
with open(yaml_path, 'w') as f:
    yaml.safe_dump(data, f)

print("Values smaller than 1e-4 have been set to 0.")
```
